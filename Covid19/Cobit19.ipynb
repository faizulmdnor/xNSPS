```python
import pandas as pd
```

```python
def drop_rename_group_df(df):
    df.drop(columns=['Province/State','Lat', 'Long'], inplace=True)
    df.groupby('Country/Region').sum()
    df.rename(columns={'Country/Region':'Country'}, inplace=True)
    df.reset_index(drop=True, inplace=True)
    return df
```

```python
df_death = pd.read_csv('time_series_covid19_deaths_global.csv')
df_recovered = pd.read_csv('time_series_covid19_recovered_global.csv')
df_confirmed = pd.read_csv('time_series_covid19_confirmed_global.csv')
```

```python
print(f"df_confirmed number of line before: {len(df_confirmed)}")
print(f"df_recovered number of line before: {len(df_recovered)}")
print(f"df_death number of line before: {len(df_death)}")
```

```python
df_confirmed = drop_rename_group_df(df_confirmed)
print(f"df_confirmed number of line after: {len(df_confirmed)}")
df_recovered = drop_rename_group_df(df_recovered)
print(f"df_recovered number of line after: {len(df_recovered)}")
df_death = drop_rename_group_df(df_death)
print(f"df_death number of line after: {len(df_death)}")
```
